# DATA CLEANSÄ°NG AND PROCESSÄ°NG

```python
# total_bedrooms verisinin medyanÄ±nÄ± aldÄ±k ve null total_bedrooms verisine atadÄ±k.
# We took the median of total_bedrooms and assigned it to the null total_bedrooms data.

df["total_bedrooms"] = df["total_bedrooms"].fillna(df["total_bedrooms"].median())
```

```python
df.isna().sum()
```

```python
sns.scatterplot(x='total_bedrooms', y='households', data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df["rooms_per_household"] = df["total_rooms"] / df["households"]
df["bedrooms_per_room"] = df["total_bedrooms"] / df["total_rooms"]
df["population_per_household"] = df["population"] / df["households"]
```

```python
df.head()
```

```python
df = df.drop(["total_rooms","total_bedrooms","population","households"], axis=1)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# TRAIN TEST SPLIT

```python
X = df.drop(["median_house_value","ocean_proximity"],axis=1)
y = df["median_house_value"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=9)
```

# REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
test_pred = regression.predict(X_test)
test_pred[:10]
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

print("Theta :     ", round(regression.coef_[0],2))
print("Intercept : ", round(regression.intercept_,2))
print("R^2 :       ", round(r2_score(y_test,test_pred),2))
print("MAE :       ", round(mean_absolute_error(y_test,test_pred),2))
print("RMSE :      ", round(np.sqrt(mean_squared_error(y_test,test_pred)),2))
```

```python
mae = mean_absolute_error(y_test, test_pred)
mse = mean_squared_error(y_test, test_pred)
score = r2_score(y_test, test_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", score)
plt.scatter(y_test, test_pred)
plt.show()
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train,y_train)
lasso_test_pred = lasso.predict(X_test)
mae = mean_absolute_error(y_test, lasso_test_pred)
mse = mean_squared_error(y_test, lasso_test_pred)
score = r2_score(y_test, lasso_test_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", score)
plt.scatter(y_test, lasso_test_pred)
plt.show()
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train,y_train)
ridge_test_pred = ridge.predict(X_test)
mae = mean_absolute_error(y_test, ridge_test_pred)
mse = mean_squared_error(y_test, ridge_test_pred)
score = r2_score(y_test, ridge_test_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", score)
plt.scatter(y_test, ridge_test_pred)
plt.show()
```

```python
from sklearn.linear_model import ElasticNet
elastic = ElasticNet()
elastic.fit(X_train,y_train)
elastic_test_pred = elastic.predict(X_test)
mae = mean_absolute_error(y_test, elastic_test_pred)
mse = mean_squared_error(y_test, elastic_test_pred)
score = r2_score(y_test, elastic_test_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", score)
plt.scatter(y_test,elastic_test_pred)
plt.show()
```
